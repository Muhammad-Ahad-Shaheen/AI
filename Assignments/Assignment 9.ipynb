```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('heart_2020_cleaned.csv')
```

```python
df['BMI'].mean()
```

```python
df['HeartDisease'].value_counts()
```

```python
df['BMI'].max(), df['BMI'].min()
```

```python
(df['BMI'] > 30).sum()
```

```python
df[(df['PhysicalActivity'] == 'No') & (df['HeartDisease'] == 'Yes')].shape[0]
```

```python
plt.hist(df['BMI'], bins=10)
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('BMI Distribution')
plt.show()
```

```python
df.groupby('PhysicalActivity')['HeartDisease'] \
  .value_counts(normalize=True)
```

```python
df['Smoking'].value_counts().plot(
    kind='pie', autopct='%1.1f%%'
)
plt.ylabel('')
plt.title('Smoking Status Distribution')
plt.show()
```

```python
df['Smoking'].value_counts(normalize=True) * 100
```

```python
plt.scatter(df['BMI'], df['SleepTime'])
plt.xlabel('BMI')
plt.ylabel('Sleep Time')
plt.title('BMI vs Sleep Time')
plt.show()
```

```python
df.boxplot(column='SleepTime')
plt.title('Sleep Time Distribution')
plt.show()
```
